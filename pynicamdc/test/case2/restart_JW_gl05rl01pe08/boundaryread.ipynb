```python
import struct
import numpy as np
import pandas as pd

file_path = "restart_all_GL05RL01z40.pe000001"  # Correct file path

with open(file_path, "rb") as f:
    # Read the first 64 bytes (Header, Characters)
    header = f.read(64).decode(errors="ignore").strip()
    
    # Read the next 256 bytes (Note, Characters)
    note = f.read(256).decode(errors="ignore").strip()
    
    # Read the next 6 integers (4 bytes each) **before rgnid**
    fmode, endiantype, grid_topology, glevel, rlevel, num_of_rgn = struct.unpack(">6I", f.read(4 * 6))

    # Read the `rgnid` array correctly (num_of_rgn elements, each 4 bytes)
    rgnid_format = f">{num_of_rgn}I"  # Struct format dynamically based on num_of_rgn
    rgnid = struct.unpack(rgnid_format, f.read(4 * num_of_rgn))  # Read rgnid array
    
    # Read `num_of_data` correctly after `rgnid`
    num_of_data = struct.unpack(">I", f.read(4))[0]  # Read 4-byte integer

    # List to store extracted data
    data_entries = []

    # Loop through each data entry
    for i in range(num_of_data):
        # Read fixed-size metadata fields
        varname = f.read(16).decode(errors="ignore").strip()
        description = f.read(64).decode(errors="ignore").strip()
        unit = f.read(16).decode(errors="ignore").strip()
        layername = f.read(16).decode(errors="ignore").strip()
        note = f.read(256).decode(errors="ignore").strip()

        # Read integer fields
        datasize, datatype, nlayer, nstep = struct.unpack(">Q3I", f.read(8 + 4 * 3))  # Q = int64, I = int32

        # Read time information
        time_start, time_end = struct.unpack(">QQ", f.read(8 * 2))  # Q = int64

        # Determine the correct data format for the data array
        dtype_map = {0: "f4", 1: "f8", 2: "i4", 3: "i8"}  # Mapping from datatype code to NumPy dtype
        data_format = dtype_map.get(datatype, None)

        if data_format:
            # Read the data array
            total_elements = datasize * nlayer * nstep
            data = np.frombuffer(f.read(total_elements * np.dtype(data_format).itemsize), dtype=np.dtype(">" + data_format))  # Big-Endian
            data = data.reshape((datasize, nlayer, nstep))  # Reshape based on metadata
        else:
            data = None  # Invalid datatype

        # Store extracted entry
        data_entries.append({
            "varname": varname,
            "description": description,
            "unit": unit,
            "layername": layername,
            "note": note,
            "datasize": datasize,
            "datatype": datatype,
            "nlayer": nlayer,
            "nstep": nstep,
            "time_start": time_start,
            "time_end": time_end,
            "data": data
        })

# Convert to DataFrame (excluding large `data` arrays for readability)
df = pd.DataFrame([{k: v for k, v in entry.items() if k != "data"} for entry in data_entries])

# Display extracted data table
import ace_tools as tools
tools.display_dataframe_to_user(name="Extracted Data Entries", dataframe=df)
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the JSON file path
json_file_path = "unit_sphere_data.json"  # Update this path if needed

# Load the JSON file
with open(json_file_path, "r") as json_file:
    loaded_data = json.load(json_file)

# Convert lists back to NumPy arrays
loaded_data_arrays = {varname: np.array(data) for varname, data in loaded_data.items()}

# Define the undefined value (-9.9999e30)
undef_value = -9.9999e30

# Extract the correct datasets for visualization
x_set1 = loaded_data_arrays[list(loaded_data_arrays.keys())[0]]  # 1st entry: X
y_set1 = loaded_data_arrays[list(loaded_data_arrays.keys())[1]]  # 2nd entry: Y
z_set1 = loaded_data_arrays[list(loaded_data_arrays.keys())[2]]  # 3rd entry: Z

x_set2 = loaded_data_arrays[list(loaded_data_arrays.keys())[3]]  # 4th entry: X
y_set2 = loaded_data_arrays[list(loaded_data_arrays.keys())[5]]  # 6th entry: Y
z_set2 = loaded_data_arrays[list(loaded_data_arrays.keys())[7]]  # 8th entry: Z

x_set3 = loaded_data_arrays[list(loaded_data_arrays.keys())[4]]  # 5th entry: X
y_set3 = loaded_data_arrays[list(loaded_data_arrays.keys())[6]]  # 7th entry: Y
z_set3 = loaded_data_arrays[list(loaded_data_arrays.keys())[8]]  # 9th entry: Z

# Mask undefined values (-9.9999e30) before visualization
masked_x_set1 = np.ma.masked_where(x_set1 == undef_value, x_set1)
masked_y_set1 = np.ma.masked_where(y_set1 == undef_value, y_set1)
masked_z_set1 = np.ma.masked_where(z_set1 == undef_value, z_set1)

masked_x_set2 = np.ma.masked_where(x_set2 == undef_value, x_set2)
masked_y_set2 = np.ma.masked_where(y_set2 == undef_value, y_set2)
masked_z_set2 = np.ma.masked_where(z_set2 == undef_value, z_set2)

masked_x_set3 = np.ma.masked_where(x_set3 == undef_value, x_set3)
masked_y_set3 = np.ma.masked_where(y_set3 == undef_value, y_set3)
masked_z_set3 = np.ma.masked_where(z_set3 == undef_value, z_set3)

# Create a 3D scatter plot for all three sets
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111, projection='3d')

# Plot each set with different colors
ax.scatter(masked_x_set1.compressed(), masked_y_set1.compressed(), masked_z_set1.compressed(), 
           c='r', s=10, label="Set 1 (Entries 1,2,3)")
ax.scatter(masked_x_set2.compressed(), masked_y_set2.compressed(), masked_z_set2.compressed(), 
           c='g', s=10, label="Set 2 (Entries 4,6,8)")
ax.scatter(masked_x_set3.compressed(), masked_y_set3.compressed(), masked_z_set3.compressed(), 
           c='b', s=10, label="Set 3 (Entries 5,7,9)")

# Labels and appearance
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("3D Scatter Plot of Cartesian Grid Locations on Unit Sphere")
ax.legend()

# Show the plot
plt.show()
```
